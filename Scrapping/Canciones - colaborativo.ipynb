```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import re
```

```python
url = "https://www.letras.com/playlists/370405/"
response = requests.get(url)
bool(response)
soup = BeautifulSoup(response.text, "html.parser")
```

```python
yr70=soup.find('div',attrs={'class':'songList-table'})
```

```python
canciones70=yr70.find_all('div',attrs={'class':'songList-table-songName'})
canciones=[]
for cancion in canciones70:
    #print(cancion.find('span'))
    canciones.append(cancion.get_text(strip=True))
```

```python
links70=yr70.find_all('a')
lista_links70=[]
for i in range(0,len(links70),2):
    lista_links70.append(links70[i]['href'])
```

```python
letras70=[]
for i in lista_links70:
    url = f"https://www.letras.com/{i}"
    response = requests.get(url)
    bool(response)
    soup = BeautifulSoup(response.text, "html.parser")
    letra= soup.find('div',attrs={'class':'lyric-original'})
    distinct=len(set(letra.get_text(strip=True)))
    letras70.append(distinct)
```
